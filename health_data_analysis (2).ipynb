```python
import pandas as pd
import matplotlib.pyplot as plt

# Function to analyze health data
def analyze_health_data(data):
    # Display basic statistics
    print("Basic Statistics:")
    print(data.describe())

    # Visualize health parameters
    parameter_to_analyze = input("Enter the name of the health parameter to analyze: ")
    if parameter_to_analyze in data.columns:
        # Plot a histogram
        plt.figure(figsize=(8, 6))
        plt.hist(data[parameter_to_analyze], bins=10, edgecolor='k')
        plt.title(f'Histogram of {parameter_to_analyze}')
        plt.xlabel(parameter_to_analyze)
        plt.ylabel('Frequency')
        plt.grid(True)
        plt.show()
    else:
        print(f"The '{parameter_to_analyze}' parameter is not found in the dataset.")

if __name__ == "__main__":
    # Prompt the user for the CSV file containing health data
    filename = '/Sleep_health_and_lifestyle_dataset.csv'

    try:
        # Load the user-provided health dataset
        health_data = pd.read_csv(filename)
        analyze_health_data(health_data)
    except FileNotFoundError:
        print("File not found. Please check the file path.")
    except Exception as e:
        print("An error occurred:", e)
```
